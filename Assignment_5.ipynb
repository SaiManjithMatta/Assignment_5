**2. Use pd_speech_features.csv**

**a. Perform Scaling**

**b. Apply PCA (k=3)**

**c. Use SVM to report performance**

```python
df_pd = pd.read_csv(r"gdrive/My Drive/pd_speech_features.csv")
```

```python
df_pd.head()
```

```python
df_pd.isnull().any()
```

```python
X = df_pd.drop('class',axis=1).values
Y = df_pd['class'].values
```

```python
# a. Perform Scaling

#Scaling Data
scaler = StandardScaler()
X_Scale = scaler.fit_transform(X)
```

```python
# b. Apply PCA (k=3)

# Apply PCA with k =3
pca3 = PCA(n_components=3)
principalComponents = pca3.fit_transform(X_Scale)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2','Principal Component 3'])

finalDf = pd.concat([principalDf, df_pd[['class']]], axis = 1)
finalDf.head()
```

```python
X = finalDf.drop('class',axis=1).values
Y = finalDf['class'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=0)
```

```python
# c. Use SVM to report performance

from sklearn.svm import SVC

svmClassifier = SVC()
svmClassifier.fit(X_train, Y_train)

y_pred = svmClassifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(Y_test, y_pred, zero_division=1))
print(confusion_matrix(Y_test, y_pred))
# Accuracy score
glass_acc_svc = accuracy_score(y_pred,Y_test)
print('accuracy is',glass_acc_svc)

#Calculate sihouette Score
score = metrics.silhouette_score(X_test, y_pred)
print("Sihouette Score: ",score) 
```

**4. Briefly identify the difference between PCA and LDA ?**

Ans:
These two are two linear transformation techniques/algorithms.

**PCA :** Singular Value Decomposition of the data is utilized to reduce its dimensionality and project it into a lower dimensional space. Before applying the SVD, the input data is centered but not scaled for each feature. It employs either the full SVD LAPACK implementation or a randomized truncated SVD.

**LDA:** This classifier with a linear decision boundary that was created by fitting class conditional densities to data and applying Bayes' rule. If all classes have the same covariance matrix, the model fits a Gaussian density to each one.


**PCA (Principal Component Analysis):**

•	An unsupervised algorithm.

•	finds directions of maximum variance regardless of class labels (It ignores class labels).

•	When to use PCA: If in case you don’t have labels for the data and want to find the low -dimensional representation.

**LDA (Linear discriminant analysis):**

•	A supervised learning algorithm.

•	finds directions of maximum class separability.

•	When to use LDA : If in case you have labeled data and want to find a low-dimensional representation.
